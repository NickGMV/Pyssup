### A quick overview of the data


Ocean minesweepers need to clear ordinance from the seas but visibility is low and touching mines is a bad idea without the appropriate preparation.

To get around this issue they take sonar readings of objects that are suspected ordinance but this data needs to be interpreted, the data in the sonar csv is readings from confirmed cases where the readings were iether mines 'M' or rocks 'R'.

This is an ideal place to apply a categorisation model like logistic regression.



#### 1. import all libraries you need Pandas, scikit, numpy etc

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
from sklearn import metrics
```

#### 2. load in your data set

```python
# setting path as a variable allows for quicker adjusting

path = './sonar.csv'

#use header = None because our data has np column headers!
sonar = pd.read_csv(path, header =None)

sonar.head()
```

#### 3. format and clean

```python
sonar.isnull().sum().sum()
```

```python
# all data types are float except for the indicator for rock/mine sonar.dtypes.tail()
```

#### 4. encode categorical variables

```python
# there is no need with this data_set the features we want to use for the model are all numeric
```

#### 5. train test split

```python
array = sonar.values

# add feature cols way of picking field
X = array[:,:-1]
y = array[:,-1:]

y.shape

#type(y)
print(f" the features of the model look like this...{X} \n\n\n the variable we are predicting looks like this {y}")
```

```python
# now to create the train test split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

#### 6. feature scaling normalisation or standardisation (normalising is 0-1, any variables) standardising is std dependent and best for normally distributed variables) plotting histograms for fields may help you decide

```python
sonar.hist(sharex=False, sharey=False, xlabelsize=1, ylabelsize=1, figsize=(12,12))
plt.show()
```

```python
# correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(sonar.corr(), vmin=-1, vmax=1, interpolation='none')
fig.colorbar(cax)
fig.set_size_inches(10,10)
plt.show()
```

#### 7. create a logistic regression object


```python
log_reg = LogisticRegression() 
```

#### 8. fit your logistic regression

```python
log_reg.fit(X_train,y_train)
```

#### 9. test your logistic regression

```python
predictions = log_reg.predict(X_test)

print(predictions)

score =log_reg.score(X_test,y_test)
```

#### 10. Confusion matrix

```python
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt="d", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
