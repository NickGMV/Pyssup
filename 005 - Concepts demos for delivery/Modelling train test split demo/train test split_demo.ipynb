```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from IPython.display import display
import time
from IPython.display import clear_output

tt_demo = pd.read_excel('./demo_tt.xlsx')
tt_demo = tt_demo.head(20)
```

```python
tt_demo
```

```python
tt_demo.iloc[:,-1:]
```

```python
X = tt_demo[['feature 1','feature 2']]
y = tt_demo['Value']
```

```python
scores = []
for i in range(1,20):
    
    X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
    
    tt_demo['in_train'] = ~tt_demo.index.isin(X_test.index)
    tt_demo['in_test'] = tt_demo.index.isin(X_test.index)
    clear_output(wait = True)
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    print(f"train test has run {i} times")
    display(tt_demo.head(20).style.apply(lambda x: ["background-color: lightgreen"]*len(x) if x['in_train']==True else ["background-color: yellow"]*len(x),axis = 1))
    
    display(f"on test {i} the model had an r^2 score of: {lr.score(X_test,y_test)}")
    scores.append(lr.score(X_test,y_test))
    
    time.sleep(1)
    
```
