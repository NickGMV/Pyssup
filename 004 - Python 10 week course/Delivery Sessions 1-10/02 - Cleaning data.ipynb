### Navigating data

### Selecting rows and columns


There are a plethora of ways to navigate your data we will cover a few here

- dot notation
- brackets notation
- loc 
- iloc
- combined

We looked at selecting columns using dot and bracket notation (refer to week 1)

We will start here by using loc

##### Loc

```python
# loc will find all the rows that have a value in the index (this can be really handy when yur index is not numerical)
countries.loc[1:100]
```

```python
# We wil create a demo of using text indexes
demo_index = countries.set_index('couuntry')
demo_index
```

```python
# we can seach the index quickly with loc
demo_index.loc['Vietnam']
```

```python
# You can give lists
demo_index.loc[['Vietnam','Algeria']]
```

```python
# Adding a condition we can find all countries beginning with a B

# loc is a little niche notice this line is easier using conventional filters 
demo_index.loc[[x for x in demo_index.index if x[0]=='B']]
```

```python
# filter all countries beginning with B
countries[countries['couuntry'].str[0]== 'B']
```

##### iloc

### Iloc

I loc allows us to navigate the data using coordinate systems. It uses a range notation where the range we select is between the numbers either side of the colon e.g 2:4 looks at 2,3 (remember the top index is not included in Python)

Link to documentation here

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html

```python
# to pick out data points use iloc
# first set of numbers is from:to for rows
#second set of numbers is from:to for columns

countries.iloc[-10:,:]

# pandas uses similar indexing rules to lists in standard python , 
# they start at zero and negative indices work too.


# IMPORTANT in a python range the largest number is NOT included by default (Notice the 5th row isn't shown) 
```

```python
#inputting only one part or other will be interpretted as everything upto or everything from
#this takes every row up to row three and every column from the 5th column (remember 0 indices!)
countries.iloc[:3,4:]
```

```python
#negative indexing can be useful to collect the last column/ row this comes in handy when processing data for
# machine learning this is because our Y (the value to be predicted) will very often be in the last column.
countries.iloc[-1:,-1:]

# first 5 rows and last column shown for each
countries.iloc[:5,-1:]
```

### Cleaning data

#### we did EDA in our last session and found some issues to clean

- Duplicates
- Nulls
- Spelling mistakes in column names
- Bad encoding
- wrong types

```python
countries.head()
countries.dtypes
```

#### Removing duplicates

```python
# find duplicates 
countries[countries.duplicated()]
```

```python
# we can remove by a) Making a subset filtering them out
sub = countries[countries.duplicated() != True]

# # or b) using the drop_duplicates() function
# countries = countries.drop_duplicates()
```

```python
# or b) using the drop_duplicates() function
countries = countries.drop_duplicates()
```

```python
# use shape to check that number of rows has changed
countries.shape
```

#### Dealing with nulls

```python
#countries.isna()
```

```python
# find nulls
countries.isna().sum()
```

```python
# deal with them using dropna()
# look in to the options with dropna() to get best results 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
countries = countries.dropna(how = 'all')
# countries = countries.fillna(0)
countries.isna().sum()
```

#### Spelling mistakes in column names

```python
#### get access to columns using .columns
countries.columns
```

```python
# change columns using rename()
countries = countries.rename(columns = {'couuntry':'country'})
countries.head()
```

```python
countries.columns = ['country', 'child_mort', 'exports', 'health', 'imports', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp']
```

#### Bad encoding

```python
# some nulls are encoded as "::" this disrupts our code from working properly
# use replace() to fix this replace all :: with a null

countries = countries.replace('::', np.nan)
countries.head()
```

```python
countries.isna().sum()
```

```python
# once the bad encoding is removed we can dropna() etc
countries = countries.dropna()
# countries.dropna(inplace = True)
```

#### Changing data type

```python
# check our dtypes
countries.dtypes
```

```python
# change columns individually
countries['child_mort'] = countries['child_mort'].astype('float')
countries.dtypes
```

```python
# change with dictionary
countries = countries.astype({"country":"string",
                  "child_mort":"float64",
                  "exports":"float64",
                  "health":"float64",
                  "imports":"float64",
                  "total_fer":"float64",
                  "gdpp":"float64"
})


countries.dtypes
```
