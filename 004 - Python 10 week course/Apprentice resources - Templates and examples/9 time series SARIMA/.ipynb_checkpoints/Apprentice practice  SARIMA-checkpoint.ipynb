```python
import pandas as pd 
import numpy as np

# for timestamps
from datetime import datetime
from datetime import timedelta

# imports to find autocorrelation plots
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf

# imports for partial correlation plots
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_pacf

#decomposing time series
from statsmodels.tsa.seasonal import seasonal_decompose

# for plotting
import matplotlib.pyplot as plt

#for adfuller test
from statsmodels.tsa.stattools import adfuller

# for performance metrics
from sklearn.metrics import mean_squared_error
```

## Creating ARIMA models




```python
# let's get some data

path = 'data/bus.csv'
bus = pd.read_csv(path)
bus.tail()
```

## 1. Cleaning

```python
bus.columns= ['Month','riders']
```

```python
bus.head(10)

# check for nulls
bus.isnull().sum().sum()
```

```python
bus.isnull().sum()
```

```python
# drop by selecting everything but last row
bus = bus.iloc[:110,:]
bus.isnull().sum().sum()
```

```python
# No nulls left
```

```python
# convert the riders column to numerical
# notice no use of lambda here! we can give a function name minus paranthesis and apply will apply it to our data
# for more complex changes write your own function
bus['riders'] = bus['riders'].apply(int)
bus.dtypes
```

## 2. Set time as index 

```python
bus['Month'] = pd.to_datetime(bus['Month'])
```

```python
bus.set_index('Month', inplace=True)
```

```python
bus.head()
```

```python
bus.isnull().sum().sum()
```

```python
bus = bus.resample("M").pad()
#bus.head()
```

## 3. Plot your timeseries to see if it has a trend or seasonality

```python
bus.plot();
```

```python
# definite trend, possible seasonality,homoscedastic
```

## 4. Decompose


REMEMBER if you don't have seasonality decompose by finding the MA and subtracting it from the data, OR by fitting a polynomial line and subtracting it this will gove you the residual.

```python
decomposition = seasonal_decompose(bus['riders'], period = 12)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(12, 6)

plt.show();


decomposition.resid.plot();
```

## 5. Acheive stationarity

```python
stationary_test = adfuller(bus['riders'])
# looks like we have a stationary timeseries as the p-value is < 0.05

stationary_test[1]
```

```python
plot_acf(bus['riders'],lags=30);
plot_pacf(bus['riders'],lags = 30);
```

```python
bus['seasonal_dif'] = bus['riders'].diff(12)
```

```python
bus.head(13)
```

```python
stationary_test = adfuller(bus['seasonal_dif'].dropna())
# looks like we have a stationary timeseries as the p-value is < 0.05

stationary_test[1]
```

```python
plot_acf(bus['seasonal_dif'].dropna(),lags=30);
plot_pacf(bus['seasonal_dif'].dropna(),lags = 30);
```

```python
bus['diffed'] = bus['seasonal_dif'].diff()
```

```python
bus.head()
```

```python
stationary_test = adfuller(bus['diffed'].dropna())
# looks like we have a stationary timeseries as the p-value is < 0.05

stationary_test[1]
```

## 6. Train test split

```python
train_size = 72

# notice we use the transformed data(logged data) to input in to our model 
# we will have to account for this later by reversing it.
train = bus['riders'][:train_size]
test = bus['riders'].iloc[train_size:]
# note depending on which way your dates are 
# sorted this may need changing

# visualise the train-test split to check it seems reasonable
fig, ax = plt.subplots(1, 1)
train.plot(ax=ax);
test.plot(ax=ax);
plt.show();
```
