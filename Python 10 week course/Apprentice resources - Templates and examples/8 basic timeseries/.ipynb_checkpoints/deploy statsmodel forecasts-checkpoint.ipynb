```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
#read in data 
df = pd.read_csv('bus.csv')
#remove bottom 2 rows due to bad formatting
df = df.iloc[:-2,:]
#check dtypes
df.dtypes
```

```python
# check data
df.head()
```

```python
#rename columns
df.columns = ['month','riders']
#change dtypes of riders to integer
df['riders'] = df['riders'].astype('int')
df.dtypes
```

```python
# make month a datetime object
df['month'] = pd.to_datetime(df['month'])
#make dat the index of the data 
df.set_index('month', inplace = True)
```

```python
# sort by date
df.sort_index()
#set th frequency and ensure no gaps
df = df.resample("M").sum().pad()
df.tail()
```

```python
#select dates for train and test set
train = df['1960':'1965-01']
test = df['1965-01':]
```

```python
#plot and look at it
train['riders'].plot()
test['riders'].plot()
```

```python
#train a model
# You can sub any midel type you like here
f_train = test.copy()
# if you get an annoying future warnign message add "initialization_method = 'estimated'" to your forecast
fit1 = ExponentialSmoothing(train['riders'] ,seasonal_periods=12 ,trend='add', seasonal='mul',initialization_method = 'estimated').fit(smoothing_level = 0.1, smoothing_seasonal = 0.1,smoothing_trend = 0.1)
f_train['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))

plt.plot( train['riders'], label='Train')
plt.plot(test['riders'], label='Test')
plt.plot(f_train['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()

# critique the model - not included here due to rush
#You should ensure that you take the residuals and plot them and that you evaluate atleast 1 off RMSE,MAE and MSE 
```

## Deploying the model

# Method 1 - use predict()

```python
# f_real will store our real forecast


# this method uses the predict method using a date range
#fit1 = ExponentialSmoothing()
fit1 = ExponentialSmoothing(df['riders'] ,initialization_method = 'estimated',seasonal_periods=12 ,trend='add', seasonal='add').fit(
    smoothing_level = 0.1, smoothing_seasonal = 0.1,smoothing_trend = 0.1)

# input a date range for predict() that starts in your original data and ends somewhere in the future
f_real = fit1.predict(start = '1969-06-30',end = '1980-08-10')
plt.figure(figsize=(16,8))


plt.plot( df['riders'], label='data to date')
plt.plot(f_real, label='exponential smoothing')
plt.legend(loc='best')
plt.show()
```
